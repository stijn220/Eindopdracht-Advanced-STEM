```python
import pandas as pd
```

```python
df_1 = pd.read_csv("2023_PM25.csv",sep=';', encoding="ISO8859-1")
df_1.columns
```

```python
# For some odd reason the column begindatumtijd has a space in front. 
# All datasets have this
df_1[' Begindatumtijd'] = pd.to_datetime(df_1[' Begindatumtijd'])
df_1['Einddatumtijd'] = pd.to_datetime(df_1['Einddatumtijd'])
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

fig, ax2 = plt.subplots(1, 1, figsize=(12, 8))
fig.autofmt_xdate()  # Angle and align the x labels


# Plot PM25 data for two stations
ax2.plot(df_1[' Begindatumtijd'], df_1['NL49701'], label='Station 49701')
ax2.plot(df_1[' Begindatumtijd'], df_1['NL49703'], label='Station 49703')
ax2.set_title('PM25 Measurements')
ax2.set_ylabel(df_1['Eenheid'].iloc[0])  # Use the unit from the data
ax2.legend()
ax2.grid(True)

# Format dates on x-axis
date_format = DateFormatter('%Y-%m-%d')

ax2.xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```
